```python
import random as rd
from matplotlib import animation
```

```python
from matplotlib import rc
rc('animation', html='html5')
```

```python
n = 1000
sd = 0.07
ylist = [rd.gauss(0, 1) for _ in range(n)]
xlist = [rd.gauss(0, 1) for _ in range(n)]

fig, ax = plt.subplots(figsize=(4,4))
ax.set_xlim((-20, 20))
ax.set_ylim((-20, 20))

pts = sum([ax.plot([], [], '.', c='b', alpha=0.1) for _ in range(0, 100)],[])
plt.close()

def drawframe(N):
    global xlist, ylist
    for i in range(N):
        pts[i].set_data(xlist, ylist)
        xlist = [x + rd.gauss(0, sd) for x in xlist]
        ylist = [y + rd.gauss(0, sd) for y in ylist]
        
    return pts

# create animiation object and render in HTML video
anim = animation.FuncAnimation(fig, drawframe, frames=100, interval=100, blit=True)
# HTML(anim.to_html5_video())
```
