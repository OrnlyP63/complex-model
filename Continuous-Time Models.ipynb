# Continuous-Time Models

- first-order continuous-time: $\frac{dx}{dt} = F(x, t)$
- higher-order continuous-time: $\frac{d^nx}{dt^n} = F(x, t);\ n>1$

---

changes higher-order to first-order

Consider: $\frac{d^2x}{dt^2} - x\frac{dx}{dt} + x^2 = 0$

Let $$\frac{dx}{dt} = \omega$$
Implies $$\frac{d\omega}{dt} = x\omega + x^2$$

---

changes non-autonomous to  autonomous funciton

$\frac{d^2x}{dt^2} - a\cos(bt) = 0$

Let $$\frac{dx}{dt} = \omega$$

Impies $\frac{d\omega}{dt} - a\cos(bt) = 0$

Let $$\frac{d\tau}{dt} = 1;\ \tau(0) = 0$$

So $$\frac{d\omega}{dt} = a\cos(b\tau)$$

---

Euler forward method

$dx/dt = G(X):$

$$x(x+\Delta t) = x(t) + G(x(t))\Delta t$$

Consdider the continuous-time logistic growth model for $0\leq t\leq 50$ with $x(0)=0.1,\ r=0.2,\ K=1$ and $\Delta t =0.01$:

$$\frac{dx}{dt} = rx\left(1-\frac{x}{K}\right)$$

Discritize the above equation

$$x(t+\Delta t) = x(t) + rx\left(1-\frac{x}{K}\right)\Delta t$$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
r = 0.2
K = 1.0
Dt = 0.01

x = 0.1
results = [x]
t = 0.0
timesteps = [t]

while t < 50:
    x = x + r * x * (1 - x / K) * Dt
    t += Dt
    
    results.append(x)
    timesteps.append(t)
    
plt.plot(timesteps, results)
plt.show()
```

Find the equilibrium points of the following model of a simple pendulum:

$$\frac{d^2\theta}{dt^2} = -\frac{g}{L}\sin(\theta)$$

Change higher-order to first-order

$$\frac{d\theta}{dt} = \omega$$

$$\frac{d\omega}{dt} = -\frac{g}{L}\sin(\theta)$$

Find equilibrium points:

$$0 = \omega$$
$$0 = -\frac{g}{L}\sin(\theta)$$

So $\theta = i\pi;\ i=0,1,2,\ldots$

The equilibrium poits $= \{(0, i\pi)\};\ i=0,1,2,\ldots$

```python
g = 9.8
L = 1

omega_values, theta_values = np.meshgrid(np.linspace(-3, 3, 50), np.pi * np.linspace(-3, 3, 50))

theta_dot = omega_values
omega_dot = -(g/L) * np.sin(theta_values)
```

```python
plt.streamplot(omega_values, theta_values, omega_dot, theta_dot)
plt.xlabel('omega')
plt.ylabel('theta')
plt.show()
```

# Lotka-Volterra model

```python
xvalues, yvalues = np.meshgrid(np.linspace(0, 3, 50), np.linspace(0, 3, 50))
xdot = xvalues - xvalues * yvalues
ydot = - yvalues + xvalues * yvalues

plt.streamplot(xvalues, yvalues, xdot, ydot)
plt.show()
```
