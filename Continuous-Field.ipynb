## Fundamentals of Vector Calculus

- Contour  
A set of spatial positions $x$ that satisty
$$f(x) = C$$ for a scalar field $f$, where $C$ is a constant.

- Gradeint  
A gradient of a scalar field f is defined as:
$$\nabla f =
\begin{bmatrix}
\frac{\partial f}{\partial x_1}\\
\frac{\partial f}{\partial x_2}\\
\vdots\\
\frac{\partial f}{\partial x_n}\\
\end{bmatrix}$$

A gradient shows a direction and how steep a slope of position $x$

- Vector of differential operators (nabla):
$$\nabla =
\begin{bmatrix}
\frac{\partial }{\partial x_1}\\
\frac{\partial }{\partial x_2}\\
\vdots\\
\frac{\partial }{\partial x_n}\\
\end{bmatrix}$$

- Divergence  
A divergence of a vector field $v$ is defined as

$$\nabla\cdot v =
\begin{bmatrix}
\frac{\partial }{\partial x_1}\\
\frac{\partial }{\partial x_2}\\
\vdots\\
\frac{\partial }{\partial x_n}\\
\end{bmatrix}^\top
\begin{bmatrix}
v_1\\
v_2\\
\vdots\\
v_n\\
\end{bmatrix} 
= \frac{\partial v}{\partial x_1} + \frac{\partial v}{\partial x_2} + \cdots \frac{\partial v}{\partial x_n}$$

Note: focus on inter product ($\cdot$) between $\nabla$ and $v$.
It literally quantifies how much the vector field v is “diverging” from the location x.

- Laplacian  
A Laplacian of a scalar field $f$ is defined as
$$\nabla^2f = \nabla\cdot\nabla f = 
\begin{bmatrix}
\frac{\partial }{\partial x_1}\\
\frac{\partial }{\partial x_2}\\
\vdots\\
\frac{\partial }{\partial x_n}\\
\end{bmatrix}^\top
\begin{bmatrix}
\frac{\partial }{\partial x_1}\\
\frac{\partial }{\partial x_2}\\
\vdots\\
\frac{\partial }{\partial x_n}\\
\end{bmatrix}f
= \frac{\partial^2 f}{\partial x^2_1} + \frac{\partial^2 f}{\partial x^2_2} + \cdots \frac{\partial^2 f}{\partial x^2_n}$$

The Laplacian of $f$, which is the divergence of the gradient field of $f$, has a
- positive value where $f$ is concave,
- or a negative value where $f$ is convex.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

### Shape of a scalar field $$f(x, y) = xy(x-1)$$

```python
# 1D arrays
x = np.arange(-10,10,0.01)
y = np.arange(-10,10,0.01)

# Meshgrid
xvalues, yvalues = np.meshgrid(x,y)
zvalues = xvalues * yvalues * (xvalues - 1)

ax = plt.subplot(projection='3d')
ax.plot_surface(xvalues, yvalues, zvalues)
plt.show()
```

### A vector field $$v(x, y) = (−xy, x − y)$$

```python
# 1D arrays
x = np.arange(-1,1,0.1)
y = np.arange(-1,1,0.1)

# Meshgrid
X, Y = np.meshgrid(x,y)

Vx = -X*Y
Vy = X - Y

# Depict illustration
plt.figure(figsize=(10, 10))
plt.streamplot(X,Y,Vx,Vy, density=1.4, linewidth=None, color='#A23BEC')
plt.plot(-1,0,'-or')
plt.plot(1,0,'-og')
plt.title('Electromagnetic Field')
  
# Show plot with grid
plt.grid()
plt.show()
```

### A laplacian $$f(x, y) = 3x^2 - xy^3$$
Consider

$$\begin{align}
\nabla^2f &= \frac{\partial^2 f}{\partial x^2} + \frac{\partial^2 f}{\partial y^2}\\
&= 6 - 6xy\\
&= 6(1 - xy)
\end{align}$$

```python
# 1D arrays
x = np.arange(-5,5,0.05)
y = np.arange(-5,5,0.05)

# Meshgrid
xvalues, yvalues = np.meshgrid(x,y)
zvalues = 6 * (1 - xvalues * yvalues) 

ax = plt.subplot(projection='3d')
ax.plot_surface(xvalues, yvalues, zvalues)
plt.show()
```
